## The Qubit â€” A Unit of Possibility

## Why a Qubit Is Limited to 0 and 1

A common question arises early:

Why not 0, 1, 2, or 3?

The key insight is this:

> A qubit is not a number.  
> It is a **choice that has not been made yet**.

That is why a qubit begins with `0` and `1`.

## What Superposition Actually Means

Superposition does **not** mean adding more labels.

It means **not committing to a choice immediately**.

- A classical bit must choose:
  - `0` or `1`

- A qubit is allowed to **delay that choice**
  - It stores how likely each outcome is
  - Without selecting either

Superposition is not confusion.  
It is **structured uncertainty**.

Nothing more is needed at the fundamental level.

## Why We Donâ€™t Add More Values

Nature handles complexity differently.

Instead of making one unit complicated,

- One undecided choice â†’ 2 possibilities

- Three undecided choices â†’ 8 possibilities
