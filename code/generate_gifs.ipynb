```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import Image

# Figure setup
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

# Classical particle
ax1.set_xlim(0, 10)
ax1.set_ylim(-2, 2)
ax1.set_title("Classical Particle")
ax1.get_yaxis().set_visible(False)
classical_dot, = ax1.plot([], [], 'ro', markersize=10)
classical_trail, = ax1.plot([], [], 'r--', lw=1)

classical_traj_x = []
classical_traj_y = []

# Quantum particle (single particle in multiple states)
ax2.set_xlim(0, 10)
ax2.set_ylim(-2, 2)
ax2.set_title("Quantum Particle (Superposition)")
ax2.get_yaxis().set_visible(False)
num_states = 10
x_quant_base = np.linspace(0, 10, num_states)
quantum_dots, = ax2.plot([], [], 'bo', markersize=6, alpha=0.7)

# Animation function
def animate(frame):
    # Classical: single predictable particle
    x_class = frame * 0.1
    y_class = np.sin(0.3*frame)
    classical_dot.set_data([x_class], [y_class])
    classical_traj_x.append(x_class)
    classical_traj_y.append(y_class)
    classical_trail.set_data(classical_traj_x, classical_traj_y)

    # Quantum: single particle in multiple states (superposition)
    # Each frame shifts the phase so the particle appears in different positions
    y_quant = np.sin(2*x_quant_base + 0.3*frame) * np.cos(0.1*frame)
    quantum_dots.set_data(x_quant_base, y_quant)

    return classical_dot, classical_trail, quantum_dots

# Create animation
ani = animation.FuncAnimation(fig, animate, frames=200, interval=50)

# Save GIF
ani.save('classical_vs_quantum_superposition.gif', writer='pillow')
plt.close(fig)
from IPython.display import Image

# Display the GIF you just saved
Image(filename='classical_vs_quantum_superposition.gif')

```
