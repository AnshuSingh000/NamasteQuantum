# 1Ô∏è‚É£ Vectors ‚Äî The Core of Quantum States

### Intuition
- A vector is **just a direction + magnitude**.  
- Think of it as an **arrow in space**.  

- In quantum computing, a **single qubit** is a vector in a 2-dimensional complex space:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

- Here:
  - $|0\rangle$ and $|1\rangle$ are the **basis vectors** (like axes in 2D space). Any qubit state can be written as a combination of |0‚ü© and |1‚ü©:
  - $\alpha$ and $\beta$ are **complex numbers** called amplitudes, telling how much the vector points along each axis.

---

### Example: Physical Intuition
- Walk 3 steps **north** ‚Üí vector $v_1 = (0,3)$  
- Walk 4 steps **east** ‚Üí vector $v_2 = (4,0)$  

Combine them tip-to-tip:

$$
v_1 + v_2 = (0+4, 3+0) = (4,3)
$$

This is like **superposition in quantum computing**: multiple directions combined.

---

### Numerical Example
- Vector A = (2,1)  
- Vector B = (-1,3)  

Linear combination: $3A - 2B$

$$
3(2,1) - 2(-1,3) = (6,3) - (-2,6) = (6,3) + (2,-6) = (8,-3)
$$

- **Tip:** Quantum operations also **scale and combine vectors**, but usually with complex numbers.

---

### Key Idea for Quantum Computing
- **Qubit = vector in 2D complex space**  
- **n qubits = a system that can be in $2^n$ different states at the same time**

**Example:**
- 1 qubit ‚Üí 2 states: |0‚ü©, |1‚ü©  
- 2 qubits ‚Üí 4 states: |00‚ü©, |01‚ü©, |10‚ü©, |11‚ü©  
- 3 qubits ‚Üí 8 states: |000‚ü©, |001‚ü©, ‚Ä¶ , |111‚ü©  

Superposition allows the system to **exist in all these states simultaneously**.

- **Superposition = adding different ‚Äúdirections‚Äù together**  
A qubit can **point partly toward |0‚ü© and partly toward |1‚ü© at the same time**.

**Example:**
$$
|\psi\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle
$$
This means the qubit is **equally likely to be measured as 0 or 1**.

- **Normalization = total probability = 1**  
The qubit‚Äôs amplitudes must satisfy:
$$
|\alpha|^2 + |\beta|^2 = 1
$$

Intuition: Think of the qubit as an **arrow that must stay on the ‚Äúunit circle‚Äù** so that probabilities make sense.  
- $|\alpha|^2$ = probability of measuring 0  
- $|\beta|^2$ = probability of measuring 1






## Why a Qubit Has Only Fixed Basis States Despite Superposition?

Think of directions in space

Imagine you are in a room with only two walls: north and east.

You can walk north (|0‚ü©) or east (|1‚ü©).

If you walk part north and part east at the same time, you are diagonally in the room.

Even though you can go diagonal, you are still in the same 2D room.

You are not creating a new wall ‚Äî just moving along a combination of the two directions.

Now for 2 qubits (4-dimensional space)

Two qubits = 4 basis directions: |00‚ü©, |01‚ü©, |10‚ü©, |11‚ü©

Any ‚Äúsuperposition‚Äù = like moving somewhere inside a 4D box, along some combination of the corners.

Analogy:

Corner	Meaning
	00‚ü©
	01‚ü©
	10‚ü©
	11‚ü©

If you take a superposition like ( |00‚ü© + |01‚ü© ) / ‚àö2, it‚Äôs like being halfway between two corners.

You are not creating a 5th corner ‚Äî just pointing somewhere along the line connecting existing corners.

## 1Ô∏è‚É£ What is a Vector? (Very Basic)

### Intuitive Idea
A **vector** represents:
- a **magnitude** (how much)
- a **direction** (which way)

Example:
- Walking **5 steps north**
  - Magnitude = 5
  - Direction = north

So:
- Number alone ‚Üí scalar  
- Number + direction ‚Üí vector  

---

### Mathematical Representation
- 2D vector:
\[
\vec{v} = (x, y)
\]

- 3D vector:
\[
\vec{v} = (x, y, z)
\]

Each component tells **how much movement along that axis**.

## 3Ô∏è‚É£ Vector Space (Very Important)

### Simple Meaning
A **vector space** is a collection of objects where:
1. Vector addition is allowed
2. Scalar multiplication is allowed

---

### Examples

#### ‚úÖ Vector Spaces
- All 2D vectors \((x, y)\)
- All real numbers
- All 3√ó3 matrices

#### ‚ùå Not Vector Spaces
- Set {0,1}
- Natural numbers only

---

### Why {0,1} is NOT a Vector Space
Try scalar multiplication:
\[
2 \times 1 = 2
\]

But:
\[
2 \notin \{0,1\}
\]

‚ùå Not closed ‚Üí fails vector space rules

---

## 4Ô∏è‚É£ Dimension (Independent Directions)

### Intuition
**Dimension = number of independent directions**

| Space | Dimension |
|------|-----------|
| Line | 1 |
| Plane | 2 |
| Space | 3 |

---

### Meaning
- 2D space ‚Üí needs **2 independent vectors**
- 3D space ‚Üí needs **3 independent vectors**

You cannot build one using the others.

---

## 5Ô∏è‚É£ Linearly Independent Vectors

### Simple Meaning
**Independent = no vector is redundant**

Each vector must add a **new direction**.

---

### Example

‚úÖ Independent:
- (1, 0)
- (0, 1)

‚ùå Dependent:
- (1, 1)
- (2, 2)

Because:
\[
(2,2) = 2(1,1)
\]

---

### Exam-Safe Definition
Vectors are **linearly independent** if:
> No vector can be written as a linear combination of the others.

---

## 6Ô∏è‚É£ Span of a Space

### Simple Meaning
**Span = cover everything**

If using given vectors you can create **every possible vector** in the space, they span it.

---

### Example
Using:
- (1,0)
- (0,1)

You can create:
\[
(x,y)
\]

‚úî Spans 2D space

Using only:
- (1,0)

‚ùå Cannot move up/down ‚Üí does NOT span 2D

---

## 7Ô∏è‚É£ Basis Vectors (Core Concept)

### Definition
A **basis** is a set of vectors that:
1. Are linearly independent
2. Span the entire space
3. Are the **minimum number required**

---

### Dimension Rule
Number of basis vectors = **dimension of the space**

---

## 8Ô∏è‚É£ Why ‚ÄúMore Vectors‚Äù Can Be Redundant and ‚ÄúFewer Vectors‚Äù Insufficient

This idea applies **only to basis vectors**, not random vectors.

---

### Case 1: Fewer Vectors ‚Üí Insufficient

2D space with only:
- (1, 0)

‚úî Can move left‚Äìright  
‚ùå Cannot move up‚Äìdown  

‚û°Ô∏è Space not fully spanned

---

### Case 2: Exact Number ‚Üí Perfect Basis

Using:
- (1, 0)
- (0, 1)

Any vector:
\[
(x, y) = x(1,0) + y(0,1)
\]

‚úî Fully spanned  
‚úî No redundancy  

---

### Case 3: More Vectors ‚Üí Redundant

Add:
- (1,1)

But:
\[
(1,1) = (1,0) + (0,1)
\]

‚ùå Adds no new direction

---

### Summary
- Too few vectors ‚Üí insufficient  
- Exact number ‚Üí perfect basis  
- Too many vectors ‚Üí redundant
